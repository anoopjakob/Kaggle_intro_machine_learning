# Pandas : Grouping and sorting

```python
pokemon = pd.read_csv('pokemon.csv')
pokemon.head()
```

```python
# we can use the groupby() function to a lot a things with respect to 
# sorting and grouping

# also we can replicate the value counts using groupby

pokemon.type_1.value_counts()

# this is the output when using value counts
```

```python
# now trying to replicate the above using groupby

pokemon.groupby('type_1').type_1.count()

# here as we can see .. we get the same ouput as above.. only
# difference is that it not in any order but .. in above  case it was in descending
# order in terms of value counts
```

__groupby()__ created a group of reviews which allottd the same point values to the given 
pokemon. Then , for each of these groups , we grabbed the __type_1__ column and counted 
how many items it appeared. __value_counts()__ is just a shortcut to this __groupby()__ operation.


```python
# here what we have done is.. we are grouping the value by the type_1
# then we are getting the values of minimum heights of each type_1

pokemon.groupby('type_1').height.min()

# that is here bug type pokemon has minimum height of 0.1
# dragon type pokemon has the minimum height of 0.3
```

you can think of each group we generate as being a slice of our dataframe containing only data with values that match. This DataFranme is accessible to us directly usint the __apply()__ method., and we can then manipulate the data in any way we see fit.For
example|||

```python
pokemon.groupby('type_1').apply(lambda df : df.height.min())

# here this is as same as the above .. but more flexible. 
# we can specify any rules inside the lambda
```

```python
pokemon.head()
```

```python
pokemon.groupby(['generation_id', 'type_1']).apply(lambda df : df.loc[df.attack.idxmax()])

# so here we can add more fine grained control..

# here we are picking out the best attacker in type_1 and in each generation
```

another __groupby()__ method worth mentioning is __agg()__ , which lets you run a bunch of different functions on your DataFrame sumultaneously 

for egs we can generate a simple statistical summary as follows

```python
pokemon.groupby(['type_1']).attack.agg([len,min,max])

# so here we can see that there are 72 total bug type_1 s.
# and minimum attack points of a pokemon in bug category is 10
#and maximum attack points of a pokemon in bug category is 139
#
```

# Multi-indexes

In all of example above we have been working with Dataframes or series objects with a single-label index.  __groupby()__ is slightly different in the fact that , depending on the operation we run , it will sometimes result in what is called a multi-index

```python
# for egs

test_pok = pokemon.groupby(['type_1','type_2']).species.agg([len])
test_pok                                                    # absence of the square 
                                                            # bracket caused the error
```

```python
mi = test_pok.index
type(mi)

# here in the output we can see that it is given it is multi-index
```

so multi indices have several methods for dealing with their tiered structure which are absent for single - level indices. They also require two levels of labels to retrieve a value.

In general the multi-index method you will use most often is the one converting back to regular indec, using __reset_index()__ method.

```python
x = test_pok.reset_index()
x.head()
```

```python
y = x.index
type(y)

# now we can see that it shows range index
```

# Sorting

looking at the __test_pok__ we can see that grouping returns data in Index order, not in value order.This is to say , when outputing the result of a __groupby__ , the order of the rows is dependent on the values in the index, not in the data.

to get data in the order we want it in, we can sort it ourselves , The __sort_values()__ method is handy for this.

```python
# first reset the index
# then use dataframe.sort_values(by ='len')

# here my dataframe used here is not compatible with using sort_values
x.sort_values(by='len')

# i think its due to duplicate columns
# i found out the error .. it happened due to the missing [] in agg in 
# defining agg

# ERROR FIXED
```

```python
# using another csv file to study about sort_values
# wine reviews.. which is the original one in kaggle learning course
```

```python
reviews = pd.read_csv('winemag-data_first150k.csv')
reviews
```

```python
# MULTI INDEX
countries_reviewed = reviews.groupby(['country','province']).description.agg([len])
countries_reviewed
```

```python
type(countries_reviewed.index)
```

```python
countries_reviewed = countries_reviewed.reset_index()
```

```python
countries_reviewed.sort_values(by='len')
```

```python
countries_reviewed.sort_index()
# sorted in alphabetic order of index since index is string format
```

```python
# you can sort by more then one column at a time
countries_reviewed.sort_values(by=['country','len'])
# so here first argument is country 
# therfore it first sort wrt country
# then it sorts according to the len in each country
```
